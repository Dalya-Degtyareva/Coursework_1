```python
# функция для заполнения недостающих месяцев
all_dates = pd.date_range(start='2016-01-01', end='2019-08-30', freq='M').strftime("%Y-%m")
def filling_empty_months(x):
    first_month = all_dates.get_loc(x.index[0])
    for date in all_dates[first_month:]:
        if date not in x.index:
            new_row = pd.Series(data={'student_id': 0}, name=date)
            x = x.append(new_row, ignore_index=False)
    x = x.sort_index()
    return x
```

```python
# подготовка датафрейма, выборка номеров первого и последнего уроков модуля
conver = df[df['is_homework'] == True]
# c = conver.groupby(['title', 'module_number'])['lesson_number'].unique()
c = conver.groupby(['title', 'module_number', 'module_title'])['lesson_number'].unique()
c = pd.DataFrame(c)

c['first_lesson'] = None
c['last_lesson'] = None
for i, row in c.iterrows():
    row['lesson_number'] = sorted(row['lesson_number'])
    row['first_lesson'] = row['lesson_number'][0]
    row['last_lesson'] = row['lesson_number'][-1]

    
c['students_start'] = None
c['students_finish'] = None
c['students_finish_2'] = None
c['conversion'] = None

# подсчет студентов по урокам
for i, row in c.iterrows():
    select_start = conver[(conver['title'] == i[0])&(conver['module_number'] == i[1])&
                          (conver['lesson_number'] == row['first_lesson'])]
    select_finish = conver[(conver['title'] == i[0])&(conver['module_number'] == i[1])&
                        (conver['lesson_number'] == row['last_lesson'])&(conver['status'] == 'done')]
    row['students_start'] = select_start.shape[0]
    row['students_finish'] = select_finish.shape[0]
    
    
# сдвиг завершённых уроков на 1 ряд вниз
modules = []
course_titles = []
for i, row in c.iterrows():
    modules.append(i[1])
    if i[0] in course_titles:
        pass
    else:
        course_titles.append(i[0])

    if len(modules) <= 1:
        pass
    elif modules[-1] < modules[-2]:
        if len(course_titles) == 2:
            c.loc[(course_titles[0], c.loc[course_titles[0]].index[-1]), 'students_finish'] = None
        else:
            c.loc[(course_titles[-2], c.loc[course_titles[-2]].index[-1]), 'students_finish'] = None

c['students_finish_2'] = c['students_finish'].shift(1)

# 1. подсчёт конверсии
for i, row in c.iterrows():
    try:
        row['conversion'] = row['students_start'] / row['students_finish_2']
    except:
        pass

c = c.fillna(0)
c.head()
```
